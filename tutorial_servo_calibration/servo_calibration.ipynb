## Comparing actual vs "perfect" servo response

The notebook will plot the measured pulse widths (x-axis) and corresponding angles (y-axis). The plot will have two lines:

- The blue line represents the actual measured positions.
- The gray dashed line represents an approximation.

A "linear approximation" is a simplified representation of data that assumes the relationship between two variables can be represented by a straight line. The concept comes from the field of mathematics, where it's often useful to approximate complex functions with simpler ones.

By comparing the actual response to the expected linear response, you can see how much your servo deviates from the expected behavior. If the blue dots fall on the gray line, your servo behaves exactly as expected. If the blue dots deviate from the gray line, it means your servo doesn't move at a constant rate, and you might need to take this into account in your project.

```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize arrays for the actual measured pulse widths and corresponding angles
pulse_widths = np.array([1010, 1100, 1210, 1301, 1400, 1500, 1603, 1711, 1800, 1900, 1920])  # in microseconds
angles = np.array([-90, -70, -50, -30, -10, 0, 10, 30, 50, 70, 90])  # in degrees

# Initialize arrays for the expected pulse widths and corresponding angles
pulse_width_range = np.linspace(np.min(pulse_widths), np.max(pulse_widths), 500)  # in microseconds
angle_range = (pulse_width_range - 1500) * (180 / 1000)  # linear approximation, 180 degrees for 1000 microseconds change

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(pulse_widths, angles, 'o', label='Measured')
plt.plot(pulse_width_range, angle_range, '--', color='gray', label='Expected')

# Mark and label the neutral position
neutral_pulse_width = 1500
neutral_angle = 0
plt.plot(neutral_pulse_width, neutral_angle, 'ro', markersize=10)  # plot the neutral position as a red dot
plt.annotate('Neutral Position', (neutral_pulse_width, neutral_angle), textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xlabel('Pulse Width (µs)')
plt.ylabel('Angle (degrees)')
plt.title('Servo Calibration')
plt.legend()

# Display the plot
plt.show()
```
